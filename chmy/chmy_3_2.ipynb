### Рыбинскова Юлия v3.2

```python
import numpy as np
import math

def f(t):  #функция
    return float((math.sin(t))/t) 

def funcValues(N, L):
    h = float(L / N)
    return [[i * h for i in range(1, N + 2)],[f(j * h) for j in range(1, N + 2)]]

def RungeKutt(N, L, u0, results):
    h = float(L / N)
    y_i = [u0] # на первом шаге y_0 = u_0
    for i in range(N): # используем формулу для вычисления очередного y_i (*)
        y_i.append(y_i[i] + h / 2.0 * (results[1][i] + results[1][i + 1]))
    return [results[0], y_i]
```

(*)   

```python
# определяем необходимые для вычислений значения
N = 20
L = 5
```

```python
# Строим таблицу значений для f(t)
Fvalues = funcValues(N, L) 
print('\n'.join([' '.join(['{:5.3f}'.format(item) for item in row]) for row in Fvalues])) 
# первая строка - t, вторая f(t)
```

```python
# Cтроим таблицу для g(x), положив g_0 = 0
g_0 = 0
Gvalues = RungeKutt(N, L, g_0, Fvalues)
print('\n'.join([' '.join(['{:5.3f}'.format(item) for item in row]) for row in Gvalues])) 
```

```python
# Cтроим таблицу для u(x), положив u_0 = 0
u_0 = 0
Uvalues = RungeKutt(N, L, u_0, Gvalues)
print('\n'.join([' '.join(['{:5.3f}'.format(item) for item in row]) for row in Uvalues])) 
```

```python
import matplotlib.pyplot as plt
plt.plot(Uvalues[0], Uvalues[1])
plt.show()
plt.plot(Uvalues[0], Uvalues[1], 'ro')
plt.show()
```

```python
N = 40 # то же самое, но для 40 точек
Fvalues = funcValues(N, L) 
Gvalues = RungeKutt(N, L, g_0, Fvalues)
Uvalues = RungeKutt(N, L, u_0, Gvalues)
plt.plot(Uvalues[0], Uvalues[1])
plt.show()
plt.plot(Uvalues[0], Uvalues[1], 'ro')
plt.show()
```
